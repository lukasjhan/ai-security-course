```python
import sqlite3
import pandas as pd
from pandas.io import sql
import os
```

```python
base_dir = '.'
```

```python
주소록 = pd.read_excel(os.path.join(base_dir, 'data', '주소록.xlsx'))
```

```python
with sqlite3.connect(os.path.join(base_dir,'data','sqliteDB')) as con: # sqlite DB 파일이 존재하지 않는 경우 파일생성
    try:
        주소록.to_sql(name = 'CUST_INFO', con = con, index = False, if_exists='fail') 
        #if_exists : {'fail', 'replace', 'append'} default : fail
    except Exception as e:
        print(str(e))
    
    query = 'SELECT * FROM CUST_INFO'
    데이터프레임1 = pd.read_sql(query, con = con)
    
    query = 'SELECT * FROM CUST_INFO where 고객번호 = "C001"'
    데이터프레임2 = pd.read_sql(query, con = con)
```
