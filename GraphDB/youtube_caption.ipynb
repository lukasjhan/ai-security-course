```python
from pytube import YouTube
import warnings
warnings.filterwarnings(action='error')
```

```python
yt = YouTube('https://www.youtube.com/watch?v=JVlVQC8iMi0')
```

```python
yt.captions
```

```python
caption = yt.captions['ko']
```

```python
print(caption.generate_srt_captions())
```

```python
from lxml import etree, html


xml_doc = html.fromstring(caption.xml_captions[40:])
print(etree.tostring(xml_doc,pretty_print=True))
```
